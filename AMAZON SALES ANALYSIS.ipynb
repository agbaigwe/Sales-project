## AMAZON SALES ANALYSIS

The objective of this data analysis project is to gain valuable insights from the ecommerce sales data in order to optimize business strategies and drive revenue growth.

Order ID: Unique identifier for each order.

Date: Date when the order was placed.

Status: Current status of the order (e.g., Shipped, Cancelled).

Fulfilment: Indicates whether the order is fulfilled by the merchant or Amazon.

Sales Channel: The platform or channel through which the sale was made (e.g., Amazon.in).

Ship Service Level: The level of shipping service chosen for the order (e.g., Standard, Expedited).

Style: Style of the product ordered.

SKU: Stock Keeping Unit, a unique identifier for each distinct product.

Category: Category of the product.

Size: Size of the ordered item.

ASIN: Amazon Standard Identification Number, a unique identifier for products on Amazon.

Courier Status: Status of the courier used for shipping (e.g., Shipped, Cancelled).

Qty: Quantity of the product ordered.

Currency: Currency in which the transaction amount is specified (e.g., INR - Indian Rupees).

Amount: Total amount of the transaction.

Ship City: City to which the order is shipped.

Ship State: State to which the order is shipped.

Ship Postal Code: Postal code of the shipping destination.

Ship Country: Country to which the order is shipped.

Promotion IDs: IDs of any promotions applied to the order.

B2B: Indicates whether the order is a business-to-business transaction (TRUE/FALSE).

Fulfilled By: Indicates whether the order is fulfilled by the merchant or through Amazon's fulfillment services (e.g., Easy Ship).
